```
import pandas as pd
clean = pd.read_csv("drive/My Drive/Veri Setleri/eda/eda81000-85000.csv")
```

```
bert(clean["clean"].iloc[4])
```

```
## df'teki tokenize columunu str columnuna çevirme
import re
def tokenize_to_str(text):
  text= re.sub(r"'","", text)
  text= re.sub(r",","", text)
  text= re.sub(r"\[","", text)
  text= re.sub(r"\]","", text)
  return text
tokenize_to_str(clean["tokenize"].iloc[7]) ## deneme
```

```
clean["join_tokenize"]=clean["tokenize"].map(tokenize_to_str)
```

```
clean.drop(['ORG_TICKET_ID', 'SR_AREA',"SR_SUB_AREA","SR_SUBTYPE_CD","X_IBB_COUNTY"], axis=1, inplace=True)
```

```
## bert i tüm datamıza uygulayalım
def label(text):
  p=bert(text)
  return p[0]["label"]

def score(text):
  p=bert(text)
  return p[0]["score"]
```

```
%%time
clean["label_T"]=clean["tokenize"].iloc[0:100].map(label)
clean["score_t"]=clean["tokenize"].iloc[0:100].map(score)
```

```
%%time
clean["label_C"]=clean["clean"].iloc[0:100].map(label)
clean["score_C"]=clean["clean"].iloc[0:100].map(score)
```

```
%%time
clean["label_J"]=clean["join_tokenize"].iloc[0:100].map(label)
clean["score_J"]=clean["join_tokenize"].iloc[0:100].map(score)
```
